# Exercícios Práticos

Crie um dataframe único com a informação do GDP e quantidade de matrículas por nível de educação por ano.

O aumento das matrículas na educação primária, secundária e terciária acompanhou o aumento do GDP per capita nos últimos 15 anos na India?

```python
import pandas as pd
```

```python
educ = 'https://raw.githubusercontent.com/abnr/ml-data/main/SYB61_T07_Education.csv'
gdp = 'https://raw.githubusercontent.com/abnr/ml-data/main/India_World_Bank_Info.csv'
```

```python
#df_educ = pd.read_csv(educ, skiprows=1,  sep = '\t', encoding = "ISO-8859-1")
df_educ = pd.read_csv(educ, header=1,   sep = '\t', encoding='unicode_escape')
```

```python
df_gdp = pd.read_csv(gdp, sep='\t', on_bad_lines='error', skiprows=4)
```

```python
df_educ
```

```python
df_educ['Series'].unique()
```

```python
df_educ_nivel_ano = df_educ[(df_educ['Unnamed: 1'] == 'India') & ((df_educ['Series'] == 'Students enrolled in primary education (thousands)') |\
                                                                  (df_educ['Series'] == 'Students enrolled in secondary education (thousands)') |\
                                                                   (df_educ['Series'] == 'Students enrolled in tertiary education (thousands)') )]
df_educ_nivel_ano
```

```python
lista_variaveis = ['Students enrolled in primary education (thousands)', 'Students enrolled in secondary education (thousands)', 'Students enrolled in tertiary education (thousands)']
df_educ_nivel_ano = df_educ[(df_educ['Unnamed: 1'] == 'India') & (df_educ['Series'].isin(lista_variaveis))]
df_educ_nivel_ano
```

```python
df_gdp
```

```python
df_gdp[(df_gdp['Indicator Name'].str.contains('GDP')) & (~df_gdp['Indicator Name'].str.contains('% of GDP')) ]['Indicator Name'].unique()
```

```python
df_gdp_Indicator_index = df_gdp.set_index('Indicator Name')
```

```python
df_gdp_Indicator_index_GDP = df_gdp_Indicator_index.loc[['GDP per capita (current US$)']]
```

```python
df_gdp_Indicator_index_GDP_melted = df_gdp_Indicator_index.loc[['GDP per capita (current US$)']].melt(id_vars= ['Country Name', 'Country Code'])
df_gdp_Indicator_index_GDP_melted
```

```python
df_gdp_Indicator_index_GDP_melted_tratada = df_gdp_Indicator_index_GDP_melted.rename(columns={'variable' : 'Year', 'value': 'GPD per capita(USD)'})
df_gdp_Indicator_index_GDP_melted_tratada = df_gdp_Indicator_index_GDP_melted_tratada.drop(columns=['Country Name','Country Code'], index= [])
df_gdp_Indicator_index_GDP_melted_tratada
```

```python
df_educ_nivel_ano_tratada = df_educ_nivel_ano.drop(columns=['Region/Country/Area','Unnamed: 1', 'Footnotes', 'Source'])
df_educ_nivel_ano_tratada
```

```python
df_gdp_Indicator_index_GDP_melted_tratada.info()
df_gdp_Indicator_index_GDP_melted_tratada = df_gdp_Indicator_index_GDP_melted_tratada.astype({'Year': 'int64'})
df_gdp_Indicator_index_GDP_melted_tratada.info()
```

```python
df_educ_nivel_ano_tratada = df_educ_nivel_ano_tratada.pivot(columns=['Series'])
df_educ_nivel_ano_tratada
```

```python
df_total_data = df_gdp_Indicator_index_GDP_melted_tratada.merge(df_educ_nivel_ano_tratada, on='Year')
df_total_data
```

```python
df_total_data_pivotado = df_total_data.pivot(columns=['Series'],index=['Year', 'GPD per capita(USD)'])
df_total_data_pivotado
```

```python
df_total_data_pivotado.replace(',','', regex=True, inplace=True)
df_total_data_pivotado = df_total_data_pivotado.astype('int64', errors='ignore')
df_total_data_pivotado
```

```python
df_total_data_pivotado= df_total_data_pivotado.reset_index()
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
y = df_total_data_pivotado['GPD per capita(USD)']
x = df_total_data_pivotado['Year']
a = (df_total_data_pivotado['Value','Students enrolled in primary education (thousands)'].astype(float).interpolate()/100)
b = (df_total_data_pivotado['Value','Students enrolled in secondary education (thousands)']/100)
c = (df_total_data_pivotado['Value','Students enrolled in tertiary education (thousands)']/100)

plt.figure(figsize=(15,10))
plt.plot(x,y, label = 'GDP')
plt.plot(x,a, label = 'Matriculas Primário/100')
plt.plot(x,b, label = 'Matriculas Secundário/100')
plt.plot(x,c, label = 'Matriculas Terciário/100')
plt.title("Matriculas por Ano x GDP per capita" ,fontsize = 14)
plt.ylabel(r'USD ou Alunos' , fontsize = 18)
plt.xlabel(r'Anos' , fontsize = 18)
plt.legend(loc = 'upper left')
plt.show()
```
