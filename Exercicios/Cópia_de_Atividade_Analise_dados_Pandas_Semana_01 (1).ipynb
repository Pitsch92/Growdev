#Exercícios

Qual a quantidade de pessoas nas seguintes faixas etárias
- 20 anos ou menos
- 21 a 35 anos
- 36 a 50 anos
- 51 a 65 anos
- maiores de 65 anos



```python
import pandas as pd
```

```python
df_pnad = pd.read_csv('/content/dados.csv')
```

```python
soma_ate_20 = df_pnad[df_pnad['Idade']<21].sort_values(by = 'Idade')\
                                          .value_counts(sort = False)\
                                          .sum()
soma_entre_21_e_35 = df_pnad[(df_pnad['Idade']<36) & (df_pnad['Idade']>20)].sort_values(by = 'Idade')\
                                                                       .value_counts(sort = False)\
                                                                       .sum()
soma_entre_36_e_50 = df_pnad[(df_pnad['Idade']<51) & (df_pnad['Idade']>35)].sort_values(by = 'Idade')\
                                                                       .value_counts(sort = False)\
                                                                       .sum()
soma_entre_51_e_65 = df_pnad[(df_pnad['Idade']<66) & (df_pnad['Idade']>50)].sort_values(by = 'Idade')\
                                                                       .value_counts(sort = False)\
                                                                       .sum()
soma_maior_65 = df_pnad[df_pnad['Idade']>65].sort_values(by = 'Idade')\
                                            .value_counts(sort = False)\
                                            .sum()

dict_faixa_etaria = {"Numero de pessoas": [soma_ate_20, soma_entre_21_e_35,\
                                           soma_entre_36_e_50, soma_entre_51_e_65,\
                                           soma_maior_65]}
pd.DataFrame(data=dict_faixa_etaria, index=['Até 20 anos', 'Entre 21 e 35 anos', 'Entre 36 e 50 anos',\
                           'Entre 51 e 65 anos', 'Mais de 65 anos'])
```

Qual estado tem a maior relação entre Renda e Anos de Estudo?

```python
df_renda_x_estudo = df_pnad[['UF', 'Renda']].groupby(by = 'UF')\
                                            .mean()\
                                            .sort_values(by= 'Renda')\
                                            .join(df_pnad[['UF', 'Anos de Estudo']].groupby(by = 'UF')\
                                                                                   .mean()\
                                                                                   .sort_values(by= 'Anos de Estudo'))
df_renda_x_estudo['Renda/Estudo'] = df_renda_x_estudo['Renda']/df_renda_x_estudo['Anos de Estudo']
```

```python
df_renda_x_estudo['Renda/Estudo'] = df_renda_x_estudo['Renda']/df_renda_x_estudo['Anos de Estudo']
df_renda_x_estudo.tail(1)
```
